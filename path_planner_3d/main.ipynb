```python
import sys
import numpy as np
from pyqtgraph.Qt import QtCore, QtWidgets
import pyqtgraph.opengl as gl

class SphereAnim:
    def __init__(self):
        self.app = QtWidgets.QApplication(sys.argv)
        self.view = gl.GLViewWidget()
        self.view.show()
        self.view.setWindowTitle('3D Sphere Animation - PyQtGraph')
        self.view.setCameraPosition(distance=20)

        meshdata = gl.MeshData.sphere(rows=20, cols=20)
        self.sphere = gl.GLMeshItem(meshdata=meshdata, smooth=True, color=(0, 0, 1, 1), shader='shaded')
        self.view.addItem(self.sphere)

        self.pos = np.array([0., 0., 0.])
        self.speed = np.array([0.1, 0.1, 0.05])

        self.timer = QtCore.QTimer()
        self.timer.timeout.connect(self.update)
        self.timer.start(30)

    def update(self):
        self.pos += self.speed
        for i in range(3):
            if self.pos[i] > 5 or self.pos[i] < -5:
                self.speed[i] = -self.speed[i]
        self.sphere.translate(*self.speed)

    def run(self):
        sys.exit(self.app.exec_())

if __name__ == '__main__':
    demo = SphereAnim()
    demo.run()

```
